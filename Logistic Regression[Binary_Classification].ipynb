```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
dataset = pd.read_excel(r"D:\Kaggle datasets\social network ads.xlsx")
dataset.drop(columns=["EstimatedSalary"],inplace = True)
```

```python
x = dataset[["Age"]]
y = dataset["Purchased"]
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
lr =LogisticRegression()
lr.fit(x_train,y_train)
```

```python
lr.score(x_test,y_test)*100
```

```python
lr.predict([[45]])
```

```python
sns.scatterplot(x="Age",y="Purchased",data = dataset)
sns.lineplot(x="Age",y = lr.predict(x),data = dataset,c="red")
plt.show()
```
