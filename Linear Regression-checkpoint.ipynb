```python
import pandas as pd
import numpy as np

# Load dataset
dataset = pd.read_excel(r"D:\Kaggle datasets\polynomial.xlsx")

# Display the first few rows
print(dataset.head())

# Define X (feature) and y (target)
X = dataset.iloc[:, 0].values.reshape(-1, 1)  # First column as feature
y = dataset.iloc[:, 1].values  # Second column as target

from sklearn.model_selection import train_test_split

# Split data (70% training, 30% testing)
x_train, x_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

from sklearn.preprocessing import PolynomialFeatures

# Create polynomial features (let's use degree=2, change if needed)
degree = 2
poly = PolynomialFeatures(degree=degree)

# Transform X_train and X_test
x_train_poly = poly.fit_transform(x_train)
x_test_poly = poly.transform(x_test)  # Only transform, no fit!

from sklearn.linear_model import LinearRegression

# Train Linear Regression model on polynomial features
model = LinearRegression()
model.fit(x_train_poly, y_train)

print("Model trained successfully!")

# Value to predict
level_value = np.array([[1.393939]])  # Change this value if needed

# Transform input using the same polynomial transformation
level_value_poly = poly.transform(level_value)

# Predict salary
predicted_salary = model.predict(level_value_poly)

print(f"Predicted Salary for Level {level_value[0][0]}: {predicted_salary[0]}")




```
